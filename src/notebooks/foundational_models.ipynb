## Initialize and Invoke a Model

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(model="llama-3.1-8b-instant", model_provider="groq")
```

```python
response = model.invoke("What is smallest particle in the atomic structure?")

print("content: ", response.content)
print("response_metadata: ", response.response_metadata)

```

## Generate Structured Output

```python
from pydantic import BaseModel, Field
from langchain.chat_models import init_chat_model
from langchain.prompts import ChatPromptTemplate


# 1. Define the structured schema using Pydantic
class AIOutput(BaseModel):
    title: str = Field(description="Smallest Particle in Atomic Structure")
    summary: str = Field(description="One paragraph summary")
    keywords: list[str] = Field(description="List of 3-5 keywords")


# 2. Build the prompt (No format instructions needed!)
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant."),
        ("user", "Generate structured output about Smallest Particle in Atomic Structure."),
    ]
)

# 3. Initialize the model and bind the structured output schema natively
llm = init_chat_model("llama-3.1-8b-instant", model_provider="groq", temperature=0)
structured_llm = llm.with_structured_output(AIOutput)

# 4. Compose the chain
chain = prompt | structured_llm

# 5. Invoke (No format instructions to pass)
result = chain.invoke({})

# The result is a fully parsed Pydantic object
print(result.title)
print(result.keywords)

# Convert to a dict if needed
print(result.model_dump())
```
